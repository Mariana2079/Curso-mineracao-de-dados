##Preenchimento de valores ausentes


```python
import pandas as pd
import numpy as np

# Criar um DataFrame de exemplo com valores ausentes
data = {'A': [1, 2, np.nan, 4, 5],
        'B': [np.nan, 2, 3, np.nan, 5],
        'C': [1, np.nan, np.nan, 4, 5]}

df = pd.DataFrame(data)

# Exibir o DataFrame original
print("DataFrame original:")
print(df)

# Preencher valores ausentes usando um valor específico (por exemplo, 0)
df_filled = df.fillna(0)

# Exibir o DataFrame com valores ausentes preenchidos
print("\nDataFrame com valores ausentes preenchidos:")
print(df_filled)

# Preencher valores ausentes usando a média das colunas
df_mean_filled = df.fillna(df.mean())

# Exibir o DataFrame com valores ausentes preenchidos pela média das colunas
print("\nDataFrame com valores ausentes preenchidos pela média das colunas:")
print(df_mean_filled)
```

##codificação de categórico para numérico (com ordem)


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = {'Tamanho': ['Grande', 'Médio', 'Pequeno', 'Médio', 'Pequeno', 'Grande']}
df = pd.DataFrame(data)

label_encoder = LabelEncoder()
df['Tamanho_encoded'] = label_encoder.fit_transform(df['Tamanho'])
df
```

##codificação de categórico para numérico (sem ordem)




```python
import pandas as pd

data = {'Cores': ['Vermelho', 'Verde', 'Azul', 'Azul', 'Verde', 'Vermelho']}
df = pd.DataFrame(data)

df_encoded = pd.get_dummies(df, columns=['Cores'], prefix=['Cor'])


df = pd.concat([df, df_encoded], axis=1)
df
```

## Normalização da base de dados

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Dados de exemplo
data = {'Feature1': [10, 20, 30, 40, 50],
        'Feature2': [5, 15, 25, 35, 45]}

df = pd.DataFrame(data)

# Inicializar o scaler MinMax
scaler = MinMaxScaler()

# Normalização Min-Max
df_normalized = scaler.fit_transform(df)

# Converter o resultado normalizado em um DataFrame
df_normalized = pd.DataFrame(df_normalized, columns=df.columns)

print("Dados Originais:")
df
```

```python
print("\nDados Normalizados (MinMax):")
df_normalized
```

## Padronização da base de dados

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Dados de exemplo
data = {'Feature1': [10, 20, 30, 40, 50],
        'Feature2': [5, 15, 25, 35, 45]}

df = pd.DataFrame(data)

# Inicializar o scaler Standardization
scaler = StandardScaler()

# Padronização
df_standardized = scaler.fit_transform(df)

# Converter o resultado padronizado em um DataFrame
df_standardized = pd.DataFrame(df_standardized, columns=df.columns)

print("Dados Originais:")
df
```

```python
print("\nDados Padronizados (Standardization):")
df_standardized
```
