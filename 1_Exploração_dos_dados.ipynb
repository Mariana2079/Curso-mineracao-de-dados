## Carregando a base de dados IRIS


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Carregando a base de dados Iris
iris = load_iris()
data = iris.data
target = iris.target

# Criando um DataFrame a partir dos dados
df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target
df
```

```python
df.describe()
```

```python
df.info()
```

```python
# Criar um boxplot para cada feature
plt.figure(figsize=(10, 6))
sns.boxplot(df)
plt.title("Boxplots das Features no Conjunto Iris")
plt.xlabel("Features")
plt.ylabel("Valores")
plt.show()

#Dados anomalos são os triangulozinhos fora da cerca na imagem em laranja (estao muito longe dos dados normais)
```

```python
# Criando o gráfico usando o Seaborn
sns.pairplot(df, hue="target")
plt.suptitle("Gráfico das Características da Base de Dados Iris")
plt.show()
```
