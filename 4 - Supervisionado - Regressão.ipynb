# Regressão

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados
vento = np.array([10, 15, 5, 12, 8, 6, 10, 14, 9, 11])
umidade = np.array([60, 45, 75, 40, 55, 70, 63, 48, 58, 52])

# Cálculos da regressão linear
X = np.column_stack((np.ones(vento.shape[0]), vento))
y = umidade
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Plotando os pontos de dados
plt.scatter(vento, umidade, label='Dados', color='blue')

# Plotando a regressão
x_fit = np.linspace(min(vento), max(vento), 100)
y_fit = theta[0] + theta[1] * x_fit
plt.plot(x_fit, y_fit, color='red', label='Regressão Linear')

plt.xlabel('Vento (km/h)')
plt.ylabel('Umidade')
plt.legend()
plt.title('Regressão Linear de Vento vs. Umidade')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dados
temperatura = np.array([25, 20, 30, 18, 22, 28, 23, 19, 24, 21])
vento = np.array([10, 15, 5, 12, 8, 6, 10, 14, 9, 11])
umidade = np.array([60, 45, 75, 40, 55, 70, 63, 48, 58, 52])

# Cálculos da regressão linear
X = np.column_stack((temperatura, vento))
X = np.column_stack((np.ones(X.shape[0]), X))
y = umidade
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Plotando os pontos de dados
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(temperatura, vento, umidade, c='b', marker='o')

# Plotando a superfície da regressão
x_fit = np.linspace(min(temperatura), max(temperatura), 10)
y_fit = np.linspace(min(vento), max(vento), 10)
x_fit, y_fit = np.meshgrid(x_fit, y_fit)
z_fit = theta[0] + theta[1] * x_fit + theta[2] * y_fit
ax.plot_surface(x_fit, y_fit, z_fit, color='r', alpha=0.5)

ax.set_xlabel('Temperatura (°C)')
ax.set_ylabel('Vento (km/h)')
ax.set_zlabel('Umidade')
ax.set_title('Regressão Linear')

plt.show()
```
