## Detecção de Outliers


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Gerar dados aleatórios em 2D
np.random.seed(42)
n_samples = 50
X = np.random.randn(n_samples, 2)

# Introduzir um outlier
outlier = np.array([5, 5])
X = np.vstack((X, outlier))

# Criar um modelo de Isolation Forest para detecção de outliers
model = IsolationForest(contamination=0.05, random_state=42)
outliers = model.fit_predict(X)

# Criar um gráfico 2D
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='blue', label='Normal')
plt.scatter(X[outliers == -1, 0], X[outliers == -1, 1], c='red', label='Outlier')
plt.title("Detecção de Outliers com Isolation Forest em um Gráfico 2D")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.ensemble import IsolationForest

# Gerar dados aleatórios em 3D
np.random.seed(42)
n_samples = 100
n_features = 3

X = np.random.randn(n_samples, n_features)

# Introduzir outliers
X[-5:] = X[-5:] + 10

# Criar um modelo de Isolation Forest para detecção de outliers
model = IsolationForest(contamination=0.05, random_state=42)
outliers = model.fit_predict(X)

# Criar uma figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotar os pontos normais em azul
ax.scatter(X[outliers == 1][:, 0], X[outliers == 1][:, 1], X[outliers == 1][:, 2], c='blue', label='Normal')

# Plotar os outliers em vermelho
ax.scatter(X[outliers == -1][:, 0], X[outliers == -1][:, 1], X[outliers == -1][:, 2], c='red', label='Outlier')

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')

plt.title("Detecção de Outliers em um Gráfico 3D")
plt.legend()
plt.show()
```
